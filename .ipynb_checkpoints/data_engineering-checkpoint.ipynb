```python
import pandas as pd
```

```python
#read the csv file into dataframes
hawaii_measurement = pd.read_csv('Resources/hawaii_measurements.csv')
hawaii_stations=pd.read_csv('Resources/hawaii_stations.csv')
```

```python
#check to see the data descriptions for the measurements
hawaii_mesurement.describe()
```

```python
#Calculate missing data percentage
total_rows = len(hawaii_mesurement)
prcp_rows = hawaii_mesurement['prcp'].count()

missing_data_percentage = (total_rows - prcp_rows)/total_rows*100
missing_data_percentage

#More than 5% of the data is missing, and hence not plausible to drop them.
```

```python
#Lets view the missing data
missing_data = hawaii_mesurement[pd.isnull(hawaii_mesurement).any(axis=1)]
missing_data.head()
```

```python
#data missing at random, from the describe, median is the best value to fill in without skewing the data. 
hawaii_measurement = hawaii_measurement.fillna(hawaii_measurement.median(),inplace=True)
#Fill them in and lets see how the data looks
hawaii_measurement.head()
```

```python
#check to see the data descriptions for the stations
hawaii_stations.describe()
#no missing value perfect
```

```python
#lets save the cleaned file
hawaii_measurement.to_csv('Resources/clean_hawaii_measurements.csv')
hawaii_stations.to_csv('Resources/clean_hawaii_stations.csv')

```
